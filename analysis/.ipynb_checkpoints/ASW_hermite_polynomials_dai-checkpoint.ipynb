```python
from sympy import *
import itertools
import numpy as np 
```

```python
v = Symbol('v')
N = 5
alpha=sqrt(2)
psis = []
for i in range(N):
    psis.append((2**i * factorial(i) * sqrt(pi))**(-1/2) * hermite(i,v/alpha) * exp(-v**2/alpha**2) /alpha) 
```

```python
def matrixA(n, T):
    # The Gram infinite matrix (truncated by n)
    var   = 1e-10   # small variation
    # param  n: the dimension of the truncated Gram matrix
    # return A: n \times n  matrix (NumPy Array)
    A = np.zeros((n, n))
    A[0,0]= 1/np.sqrt(2*T) 
    for i in range(1, n):
        A[i, i]  = A[i-1, i-1]*(2*i-1)/(2*i)

    for i in range(0, n):
        for j in range(1, i+1):  ## changement Ã  vÃ©rifier
            if (i-j) >= 0 and (i+j) < n:
                A[i-j, i+j]  = -np.sqrt( (i-(j-1))/(i+j) ) * A[i-(j-1), i+(j-1)]
                A[i+j, i-j]  =  A[i-j, i+j]

    for i in range(0, n):
        A[i, i] += var 

    return A
```

```python
A_french = matrixA(50, 2)
```

```python
A = zeros(N,N)
for (row,column) in itertools.product(range(N),range(N)):
    A[row,column] = integrate(psis[row]*psis[column], (v,-oo,oo))
```

```python
A.evalf(8,chop=True)
```

```python
dpsis = []
for i in range(N):
    dpsis.append(diff(psis[i],v)) 

D = zeros(N,N)
for (row,column) in itertools.product(range(N),range(N)):
    D[row,column] = sqrt(2*row) * KroneckerDelta(row-1,column)
```

```python
D.evalf(4,chop=True)
```

```python
(A@D+D.T@A).evalf(2,chop=True)
```

```python
n = 4 # truncated matricies, n<N
A11 = A[0:n,0:n]
A12 = A[0:n,n:]
D21 = D[n:,0:n]
D11 = D[0:n,0:n]
```

```python
D21.evalf(2, chop=True)
```

```python
D11.evalf(2,chop=True)
```

```python
# method 1, fix the diffentiation matrix by correcting it with higher moment terms
Dmethod1 = D11 + A11.inv()@A12@D21
Dmethod1.evalf(2,chop=True)
```

```python
(A11@Dmethod1 + Dmethod1.T@A11).evalf(10,chop=True)
```

```python
# method 2, penalize nonideal behavior
eps = 1e-14
Dmethod2 = 0.5*D11 - 0.5*(A11+eps*eye(n)).inv()@D11.T*(A11+eps*eye(n))
Dmethod2.evalf(2,chop=True)
```

```python
(A11@Dmethod2 + Dmethod2.T@A11).evalf(10,chop=True)
```
