# Lets scan for optimal $\alpha$

```python
alpha_vec = np.linspace(0.5, 2, 80)
Nv_vec = np.arange(10, 100, 10)
alpha_new_vec = np.zeros(len(Nv_vec))
```

```python
Nv_vec
```

```python
error_vec = np.zeros((len(alpha_vec), len(Nv_vec)))
```

```python
for jj in range(len(alpha_vec)):
    for kk in range(len(Nv_vec)):
        C = np.zeros(Nv_vec[kk]) 

        for ii in range(Nv_vec[kk]):
            C[ii] = scipy.integrate.trapezoid(f0*aw_psi_hermite_complement(n=ii, alpha_s=alpha, u_s=u, v=v), x=v, dx=np.abs(v[1] - v[0])) /alpha

        alpha_new_vec[kk] = updated_alpha(alpha_prev=alpha, C20=C[2], C10=C[1], C00=C[0])
        P = P_case_iii(alpha_prev=alpha, alpha_curr=alpha_vec[jj], Nv=Nv_vec[kk])
        C_new = P @ C
        sol = np.zeros(len(v))
        for nn in range(Nv_vec[kk]):
            sol += C_new[nn] * aw_psi_hermite(n=nn, alpha_s=alpha_vec[jj], u_s=u, v=v)
        error_vec[jj, kk] = np.linalg.norm(np.abs(sol-f0), ord=2)
```

```python
fig, ax = plt.subplots(figsize=(6, 3))
for kk in np.arange(0, len(Nv_vec), 2):
    ax.plot(alpha_vec, error_vec[:, kk], label="$N_{v} = $" + str(Nv_vec[kk]))
    ax.scatter(alpha_vec[np.argmin(error_vec[:, kk])], np.min(error_vec[:, kk]), marker="x", s=50)

ax.axvline(alpha_new_vec[kk], color="black", ls="--", label="physics-based")
legend = ax.legend(ncols=2, fancybox=False, shadow=False, fontsize=12, loc='lower right')
legend.get_frame().set_alpha(0)
ax.set_ylim(1e-3, 4)
ax.set_xlim(0.5, 2)
ax.set_yscale("log")
ax.set_xlabel(r"$\alpha$")
ax.set_ylabel(r"$L_{2}$ error")
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

plt.tight_layout()
plt.savefig("../figures/functional_analysis/alpha_optimal.png", dpi=300) 
```

# Is my P matrix kosher?

```python
P =  P_case_iii(alpha_curr=alpha_new, alpha_prev=alpha, Nv=Nv)
```

```python
def proj_1d_Koshkarov(a,b,Nn):
    Pnn = np.zeros((Nn,Nn))
    for n in range(Nn):
        for m in range(Nn):
            if n == m:
                Pnn[n,m] = 1/a**(n+1)
                
            elif n > m: # lower diagonal
                
                if (abs(b) < 1e-16 and abs(a - 1.0) < 1e-16):
                    return Pnn # return 1 on diagonal?
                #elif (abs(b) < 1e-16):
                elif (b == 0.0):

                    el = 1/a
                    if ((n-m)%2 != 0):
                        continue
                    for i in range(2, n-m+1, 2):
                        el *= np.sqrt((i-1.0)/(1.0*i))*(1.0/(a*a) - 1.0)
                    for j in range(1, m+1):
                        el *= np.sqrt((1.0*(j+(n-m)))/(1.0*j))/a
                    Pnn[n,m] = el
                # elif (abs(a - 1.0) < 1e-16):
                elif (a == 1.0):
                    el = 1.0
                    
                    for i in range(1, n+1):
                        el *= -np.sqrt(2.0/i)*b

                    for j in range(1, m+1):
                        el *= -1.0/(b*np.sqrt(2*j)) * (n+1-j)
                    Pnn[n,m] = el
                else:
                    alpha = -2.0*b/a
                    beta  = 1.0/(a*a) - 1.0
                    chi   = beta/(alpha*alpha)
                    
                    temp = 1.0
                    for k in range(1, n-m+1):
                        temp *= alpha/k
                    
                    sum = temp
                    for k in range(1, int((1e-16+1.0*(n-m))/2.0) + 1):
                    # for k in range(1, (n-m)/2+1):
                        temp *= (n-m-2*k+2)*(n-m-2*k+1)*chi/k
                        sum += temp
                    
                    sum *= 1.0/a**(m+1)
                    for k in range(m+1, n+1):
                        sum *= np.sqrt(k/2.0)
                        
                    Pnn[n,m] = sum
    return Pnn
```

```python
P_koshkarov = proj_1d_Koshkarov(a= a_constant(alpha_curr=alpha_new, alpha_prev=alpha),
                                b=b_constant(u_curr=u, u_prev=u, alpha_prev=alpha),Nn=Nv)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(10, 3))
pos = ax[0].imshow(P_koshkarov)
cbar = fig.colorbar(pos)
ax[0].set_title("Alex's")
pos = ax[1].imshow(P)
cbar = fig.colorbar(pos)
ax[1].set_title("Opal's")
pos = ax[2].imshow(np.abs(P_koshkarov-P))
ax[2].set_title("abs. difference")
cbar = fig.colorbar(pos)
plt.tight_layout()
plt.savefig("../figures/functional_analysis/P_approx.png", dpi=300) 
```
