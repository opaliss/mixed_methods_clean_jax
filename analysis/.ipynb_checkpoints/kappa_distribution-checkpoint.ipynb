```python
def kappa(v, k, T=1):
    return (1 + v**2)**(-(k+1)) * scipy.special.gamma(k+1)/scipy.special.gamma(k-0.5) / np.sqrt(k**3) /np.sqrt(2*np.pi)
```

```python
x = np.linspace(-10, 10, 1000)


maxwellian_vec = np.exp(-x**2)/np.sqrt(2*np.pi)

for ii in range(1, 100, 5):
    kappa_vec = kappa(v=x, k=ii)
    plt.plot(x, kappa_vec, label="kappa $\kappa =$" + str(ii))
plt.plot(x, maxwellian_vec, label="Maxwellian")
plt.legend()
```
