```python
%matplotlib inline
import matplotlib.pyplot as plt
from mycolorpy import colorlist as mcp
import matplotlib
font = {'family' : 'serif',
        'size'   : 14}

matplotlib.rc('font', **font)
matplotlib.rc('xtick', labelsize=14) 
matplotlib.rc('ytick', labelsize=14) 

# Enable LaTeX rendering
matplotlib.rcParams['text.usetex'] = True
```

```python
def get_L(NH, alpha, u, k):
    L = np.zeros((NH, NH), dtype="complex128")
    for nn in range(NH):
        L[nn, nn] += u * 1j * k 
        L[nn, nn-1] += alpha * np.sqrt(nn/2) * 1j * k
        if nn != NH - 1:
            L[nn, nn+1] += alpha * np.sqrt((nn+1)/2) * 1j * k 
    return L
```

```python
def get_M(NH, alpha, u, k, dalphadx, dudx):
    M = np.zeros((NH, NH), dtype="complex128")
    for nn in range(NH):
        # u terms 
        M[nn, nn-2] += dudx * np.sqrt(nn * (nn-1)) 
        M[nn, nn] += dudx * (nn+1)
        M[nn, nn-1] += u/alpha * dudx * np.sqrt(2*nn) 
        # alpha terms 
        M[nn, nn-3] += dalphadx * np.sqrt(nn*(nn-1)*(nn-2)/2) 
        M[nn, nn-2] += dalphadx * u / alpha * np.sqrt(nn*(nn-1))
        M[nn, nn] += dalphadx * u /alpha * (nn+1) 
        if nn != NH - 1:
            M[nn, nn+1] += dalphadx * (nn+2) * np.sqrt((nn+1)/2)
        M[nn, nn-1] += dalphadx * (2*nn+1) * np.sqrt(nn/2) 
    return M
```

```python
def get_N(NH, alpha, dphidx):
    N = np.zeros((NH, NH), dtype="complex128")
    for nn in range(NH):
        N[nn, nn-1] = - np.sqrt(2*nn)/alpha * dphidx
    return N
```

```python
alpha = 1
u = 0 
k = 1
dudx = np.linspace(-2, 2, 200)
dalphadx = np.linspace(-3, 3, 100)
dphidx = 1
NH = 100
```

```python
A_total = np.zeros((len(dudx), len(dalphadx)))
```

```python
for ii in range(len(dudx)):
    for jj in range(len(dalphadx)):
        O = - get_L(NH=NH, alpha=alpha, u=u, k=k) \
            - get_M(NH=NH, alpha=alpha, u=u, k=k, dalphadx=dalphadx[jj], dudx=dudx[ii]) \
            - get_N(NH=NH, alpha=alpha, dphidx=dphidx)
        A_total[ii, jj] = np.max((np.linalg.eig(O)[0]).real)
```

```python
fig, ax = plt.subplots()
pos = ax.contourf( dudx, dalphadx, A_total.T, levels=10)
cbar = fig.colorbar(pos)
ax.scatter(0, 0, marker="x", color="red")
ax.set_ylabel(r"$\partial \alpha/\partial x$")
ax.set_xlabel(r"$\partial u/\partial x$")
ax.set_title(r"$\alpha = 1, u =0, \partial \phi/ \partial x = 1, k=1, N_{H} = $" + str(NH))
cbar.ax.set_ylabel(r"$\mathrm{max}(\mathrm{Re}(\lambda))$")
plt.tight_layout()
plt.savefig("../figures/adaptive/stability_plot_" + str(NH) + ".png", dpi=400)
```
